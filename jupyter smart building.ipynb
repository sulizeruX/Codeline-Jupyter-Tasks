# Smart Building

## 1.

```python
import numpy as np

time_in_minutes = np.arange(0, 1440)
print(time_in_minutes)
```

```python
temperature_data = np.random.normal(loc=22, scale=5, size=1440)
print(temperature_data)
```

```python
humidity_data = np.random.normal(loc=55, scale=15, size=1440)
print(humidity_data)
```

```python
pressure_data = np.random.normal(loc=1012, scale=10, size=1440)
print(pressure_data)
```

```python
home_data = np.concatenate((time_in_minutes.reshape(-1, 1),
                            temperature_data.reshape(-1, 1),
                            humidity_data.reshape(-1, 1),
                            pressure_data.reshape(-1, 1)), axis=1)
print(home_data)
```

## 2.

```python
average_temperature = home_data[:, 1].mean()
print(f"Average Temperature: {average_temperature}")
average_humidity = home_data[:, 2].mean()
print(f"Average Humidity: {average_humidity}")
average_pressure = home_data[:, 3].mean()
print(f"Average Pressure: {average_pressure}")

max_temperature = home_data[:, 1].max()
print(f"Max Temperature: {max_temperature}")
min_temperature = home_data[:, 1].min()
print(f"Min Temperature: {min_temperature}")
max_humidity = home_data[:, 2].max()
print(f"Max Humidity: {max_humidity}")
min_humidity = home_data[:, 2].min()
print(f"Min Humidity: {min_humidity}")
max_pressure = home_data[:, 3].max()
print(f"Max Pressure: {max_pressure}")
min_pressure = home_data[:, 3].min()
print(f"Min Pressure: {min_pressure}")
```

```python
std_average_temperature = home_data[:, 1].std()
print(f"Standard Deviation of Average Temperature: {std_average_temperature}")
std_average_humidity = home_data[:, 2].std()
print(f"Standard Deviation of Average Humidity: {std_average_humidity}")
std_average_pressure = home_data[:, 3].std()
print(f"Standard Deviation of Average Pressure: {std_average_pressure}")

std_max_temperature = home_data[:, 1].std()
print(f"Standard Deviation of Max Temperature: {std_max_temperature}")
std_min_temperature = home_data[:, 1].std()
print(f"Standard Deviation of Min Temperature: {std_min_temperature}")
std_max_humidity = home_data[:, 2].std()
print(f"Standard Deviation of Max Humidity: {std_max_humidity}")
std_min_humidity = home_data[:, 2].std()
print(f"Standard Deviation of Min Humidity: {std_min_humidity}")
std_max_pressure = home_data[:, 3].std()
print(f"Standard Deviation of Max Pressure: {std_max_pressure}")
std_min_pressure = home_data[:, 3].std()
print(f"Standard Deviation of Min Pressure: {std_min_pressure}")
```

## 3.

```python
count = 1
for i in range(0, 1440, 60):
    avg_60_min_temperature = home_data[i:i+60, 1].mean()
    avg_60_min_humidity = home_data[i:i+60, 2].mean()
    avg_60_min_pressure = home_data[i:i+60, 3].mean()
    print(f"Average for {count} hour(s): Temperature: {avg_60_min_temperature}, Humidity: {avg_60_min_humidity}, Pressure: {avg_60_min_pressure}")
    count += 1
```

## 4.

```python
for i in range(0, 1440):
    if home_data[i, 1] > 25:
        print(f"High Temperature Alert at {home_data[i, 0]} minutes: {home_data[i, 1]}°C")
```

```python
for i in range(0, 1440):
    if home_data[i, 1] == min_temperature:
        print(f"Minimum tempurature occured at {home_data[i, 0]} minutes")
```
